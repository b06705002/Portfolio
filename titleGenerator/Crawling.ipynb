```python
import json
df = []
with open("./news/news.json", "r") as jsfile:
    data = jsfile.readline()
    df.append(eval(data))
    count = 0
    while(1):
        count += 1
        try:
            data = jsfile.readline()
            df.append(eval(data))
        except:
            break
```

```python
from nltk.lm import MLE
from time import sleep
```

```python
from bs4 import BeautifulSoup
import os
import requests

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
folder = "./sportNews/"
Error_list = []
Null_list = []
last = 67845
for i in range(last+1, len(df)):
    if df[i]['category'] != "SPORTS":
        continue
    if i % 5000 == 0:
        print(i)
    url = df[i]['link']
    try:
        response = requests.get(url, headers = headers)
    except Exception as e:
        print(i)
        print(e)
        Error_list.append(i)
        continue
    soup = BeautifulSoup(response.text, "html.parser")
    sel = soup.select("div.content-list-component.yr-content-list-text.text")

    article = ""
    for text in sel:
        article += str(text.text)
    if article == "":
        print(df[i])
        Null_list.append(i)
        continue
    fpath = os.path.join(folder, "SPORT_" + str(i)+".txt")
    with open(fpath, "w", encoding = 'utf-8') as f:
        f.write(article)
```
