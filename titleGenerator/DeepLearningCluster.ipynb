```
import pickle

topic = "SPORTS"
with open('./drive/MyDrive/4rd1/DataMining/IRTM_Final/{}_raw_after_clustering.pkl'.format(topic), "rb") as fp:
  train = pickle.load(fp)
```

```
test = {i:0 for i in range(20)}
for i in train:
  test[i['k20']] += 1
print(test)
```

```
# !pip install headliner
from headliner.trainer import Trainer
from headliner.model.basic_summarizer import BasicSummarizer
from headliner.model.attention_summarizer import AttentionSummarizer
from headliner.model.transformer_summarizer import TransformerSummarizer
from headliner.model.bert_summarizer import BertSummarizer
from sklearn.model_selection import train_test_split

summarizer = TransformerSummarizer(embedding_size=64, max_prediction_len=20, embedding_encoder_trainable=True)

limit = len(train)
K = 8

X = [train[i]['content'] for i in range(limit) if train[i]['k20'] == K]
Y = [train[i]['headline'] for i in range(limit) if train[i]['k20'] == K]

x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.33, random_state=42)

data_train = [(x_train[i], y_train[i]) for i in range(len(x_train))]
data_val = [(x_val[i], y_val[i]) for i in range(len(x_val))]
```

```
# data = [('You are the stars, earth and sky for me!', 'I love you.'),
#         ('You are great, but I have other plans.', 'I like you.')]
# train summarizer and save model

def read_data_iteratively():
    return ((x_train[i], y_train[i]) for i in range(len(x_train)))

class DataIterator():
    def __iter__(self):
        return read_data_iteratively()
data_iter = DataIterator()

# basic_summarizer = BasicSummarizer()
# attention_summarizer = AttentionSummarizer()
# transformer_summarizer = TransformerSummarizer()
# bert_summarizer = BertSummarizer()
# summarizer = TransformerSummarizer(num_layers=2)
# trainer = Trainer(batch_size=2, steps_per_epoch=100)
# trainer.train(summarizer, data_train, num_epochs=20, val_data=data_val)

# summarizer = TransformerSummarizer(embedding_size=64, max_prediction_len=20, embedding_encoder_trainable=True)
# summarizer = AttentionSummarizer(max_prediction_len=20, embedding_encoder_trainable=True)
trainer = Trainer(batch_size=16, steps_per_epoch=len(x_train))
trainer.train(summarizer, data_train, num_epochs=2, val_data=data_val)
# trainer.train(summarizer, data_iter, num_epochs=5, val_data=data_val)
```

```
# summarizer = TransformerSummarizer.load('./tmp/summarizer')
obj = 5
prediction = summarizer.predict(data_train[obj][0])
print("Training True headline:\n", data_train[obj][1].lower())
print("")
print("Training Predict headline:\n", prediction.replace('<end>', ''))

prediction = summarizer.predict(data_val[obj][0])
print("\n")
print("Testing True headline:\n", data_val[obj][1].lower())
print("")
print("Testing Predict headline:\n", prediction.replace('<end>', ''))
```
